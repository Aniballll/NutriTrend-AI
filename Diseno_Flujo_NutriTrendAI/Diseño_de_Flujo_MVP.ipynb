# Proyecto: Análisis y Predicción del Consumo Alimentario en España
**Curso:** Data Science & IA  
**Estudiante:** Aníbal García García-Moreno

## 1. Flujo Funcional del MVP (Happy Path)
El objetivo del MVP es permitir que un usuario (analista de salud o ciudadano) visualice la evolución de la dieta en su región.

1.  **Selección de Parámetros:** El usuario accede al dashboard y selecciona una **Comunidad Autónoma** y una **Categoría de Alimento** (ej. Frutas frescas vs. Ultraprocesados).
2.  **Carga de Datos:** El sistema filtra las series históricas correspondientes en la base de datos.
3.  **Procesamiento Predictivo:** El modelo de IA calcula la tendencia de consumo para los próximos 12 meses basándose en el histórico.
4.  **Visualización de Resultados:** El sistema genera un mapa de calor (riesgo nutricional) y una gráfica de líneas (predicción).
5.  **Exploración:** El usuario puede comparar su región con la media nacional para obtener contexto.

## 2. Componentes Técnicos
Para la implementación de este flujo, se utilizarán los siguientes componentes:

* **Interfaz de Usuario:** Aplicación web interactiva construida con **Streamlit** (Python).
* **Lógica de Negocio (Backend):** Scripts de **Python** para el procesamiento de datos y filtrado dinámico.
* **Modelo de IA:** Algoritmos de series temporales (**Prophet** o **Exponential Smoothing**) para la parte predictiva.
* **Almacenamiento:** Archivos de datos optimizados en formato **CSV** o **Parquet**, derivados de la limpieza previa de fuentes oficiales.

A continuación se detallan los datos requeridos en cada fase del flujo:

## 4. Razonamiento del MVP
Este flujo se centra en la **democratización de los datos**. En lugar de tablas complejas, el usuario obtiene una respuesta visual clara sobre el riesgo nutricional de su entorno. Al utilizar fuentes oficiales como el INE y el MAPA, garantizamos que el análisis sea sólido y representativo de la realidad española.

## 5. Justificación de Viabilidad y Valor Estratégico

Para asegurar que el proyecto es realizable con mis conocimientos actuales y aporta un valor real, he estructurado la solución bajo los siguientes pilares:

#### **Coherencia Tecnológica y Unificación de Entorno**
* **Ecosistema 100% Python:** Al utilizar un entorno unificado, elimino la complejidad de integrar lenguajes externos (como PHP). Esto facilita el mantenimiento y la velocidad de ejecución del MVP.
* **Interfaz con Streamlit:** Selecciono esta herramienta por ser el estándar en la industria de datos, permitiendo transformar scripts de Python en aplicaciones web interactivas de forma nativa y eficiente.

#### **Solidez Técnica y Uso de Librerías Estándar**
* **Tratamiento de Datos:** El uso de **Pandas** y **Numpy** me permite realizar una limpieza profunda de los microdatos del **INE** y el **MAPA**, asegurando que el modelo trabaje con información real y de alta confianza.
* **Capacidad Predictiva:** La lógica de predicción se apoya en bibliotecas especializadas (como **Prophet** o **Scikit-learn**), lo que garantiza la fiabilidad de los cálculos sin necesidad de desarrollar algoritmos desde cero.

#### **Impacto y Escalabilidad**
* **Calidad de Fuentes:** El proyecto se basa en datos públicos y granulados, lo que reduce el riesgo de sesgos y garantiza la veracidad de los perfiles de consumo detectados.
* **Potencial de Crecimiento:** Aunque el MVP se centra en tendencias de consumo, el modelo es escalable. En fases posteriores, se podrían cruzar estos resultados con indicadores de salud (tasas de obesidad o diabetes) para crear un sistema de alerta temprana de salud pública.
