## 1. Resumen Ejecutivo del Plan
Este documento detalla la **estrategia de ejecución** para el desarrollo del MVP de **NutriTrend España**. El enfoque se centra en la integración de fuentes de datos masivas del **INE** y el **MAPA** para construir un modelo predictivo capaz de anticipar tendencias nutricionales. El proyecto se desarrollará bajo una metodología ágil, priorizando la obtención de un producto funcional en un plazo de **6 semanas**.

## 2. Desglose de Bloques de Trabajo (WBS)

###  Bloque I: Ingesta de Datos y ETL (Extracción, Transformación y Carga)
*Este es el cimiento del proyecto. Sin datos limpios y estructurados, el modelo de IA no será fiable.*

* **Tarea 1: Adquisición de Fuentes Oficiales.** Descarga de microdatos de la Encuesta de Salud (INE 2023/2025) y series históricas del Panel de Consumo (MAPA).
* **Tarea 2: Preprocesamiento con Pandas.** Limpieza de valores nulos, normalización de categorías de alimentos y unificación de códigos regionales (CC.AA.).
* **Tarea 3: Análisis Exploratorio (EDA).** Generación de estadísticas descriptivas iniciales para validar que los datos reflejan la realidad del consumo actual.

###  Bloque II: Inteligencia Artificial y Modelado Predictivo
*Transformación de datos históricos en proyecciones a futuro.*

* **Tarea 4: Implementación del Modelo de Series Temporales.** Configuración y entrenamiento de algoritmos (como **Prophet** o **Exponential Smoothing**) para modelar el consumo por región.
* **Tarea 5: Validación y Métricas.** Evaluación del rendimiento del modelo utilizando el Error Absoluto Medio (**MAE**) para asegurar que las predicciones sean precisas y útiles para salud pública.

###  Bloque III: Desarrollo de la Interfaz (Frontend & Visualización)
*Hacer que los datos sean comprensibles para cualquier usuario.*

* **Tarea 6: Arquitectura del Dashboard con Streamlit.** Diseño de la interfaz de usuario, incluyendo selectores dinámicos de región, sexo y categoría de alimento.
* **Tarea 7: Visualización Geoespacial.** Integración de mapas de calor que identifiquen visualmente los "puntos calientes" de riesgo nutricional en la geografía española.

##  4. Gestión de Riesgos y Planes de Contingencia

### **Riesgo 1: Incompatibilidad de formatos entre INE y MAPA**
* **Impacto:** Alto (Retraso en la fase de ETL).
* **Plan B:** En lugar de cruzar microdatos a nivel individual, se utilizarán tablas de indicadores agregados (proporcionadas por el INE en 2025) para simplificar la unión de datos.

### **Riesgo 2: Baja estacionalidad o datos ruidosos en ciertas regiones**
* **Impacto:** Medio (Predicciones poco precisas).
* **Plan B:** Pivotar el valor del MVP hacia un **Dashboard Descriptivo Avanzado**, donde el peso del proyecto recaiga en la comparativa histórica detallada y la identificación de patrones actuales.
