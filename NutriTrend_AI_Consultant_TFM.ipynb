```python
import pandas as pd
import os
import requests

print("INITIALIZING NUTREND AI CONSULTANT CORE (EXACT HTTP MODE)...")

# 1. Copiamos los parámetros exactos del comando curl que te da Google
API_KEY = "TU API KEY DE GOOGLE"
HEADERS = {
    "Content-Type": "application/json",
    "X-goog-api-key": API_KEY  # Cambiado al estándar de tu curl
}

# 2. Ruta y carga de la base de datos limpia
DATASET_LIMPIO = "dataset_nutritrend_limpio.csv"

if os.path.exists(DATASET_LIMPIO):
    df_ai = pd.read_csv(DATASET_LIMPIO)
    print(f"[ÉXITO] Dataset de TFM cargado correctamente.")
    print(f"{df_ai.shape[0]} Comunidades Autónomas listas para el análisis.")
    
    # Pasamos los datos a formato texto para meterlos de contexto
    contexto_datos = df_ai.to_string(index=False)
    print("Datos preparados y listos para inyección.")
else:
    print(f"[ERROR] No se encuentra '{DATASET_LIMPIO}' en esta carpeta.")
```

```python
def consultar_agente_nutritrend(pregunta_usuario):
    import requests
    global contexto_datos, HEADERS
    
    # Instrucciones de rol ejecutivo para tu tribunal
    instrucciones = (
        "Actúa como un Consultor Experto en Salud Pública, Nutrición y Economía de España.\n"
        f"Analiza de forma rigurosa este dataset de las 17 Comunidades Autónomas:\n\n{contexto_datos}\n\n"
        "Reglas de comportamiento:\n"
        "1. Tus respuestas deben ser muy breves, ejecutivas y concisas.\n"
        "2. Usa los datos numéricos de renta y consumo del dataset para argumentar por qué el sur/centro (como Extremadura o Castilla-La Mancha) sale clasificado en Riesgo Alto por el modelo K-Means."
    )
    
    # Construcción del cuerpo del JSON respetando los niveles de tu ejemplo 'parts' y 'contents'
    payload = {
        "contents": [
            {
                "parts": [
                    {
                        "text": f"{instrucciones}\n\nPREGUNTA DEL INVESTIGADOR: {pregunta_usuario}"
                    }
                ]
            }
        ]
    }
    
    # URL EXACTA de tu comando curl de Google AI Studio
    url = "https://generativelanguage.googleapis.com/v1beta/models/gemini-flash-latest:generateContent"
    
    try:
        res = requests.post(url, headers=HEADERS, json=payload, timeout=15)
        
        if res.status_code == 200:
            # Extracción del texto del JSON de respuesta
            respuesta_texto = res.json()['candidates'][0]['content']['parts'][0]['text']
            return respuesta_texto
        else:
            return f"Error de la API de Google ({res.status_code}): {res.text}"
            
    except Exception as e:
        return f"Error crítico de conexión: {str(e)}"

print("Motor HTTP oficial configurado. El Consultor de IA está listo.")
```

#### 2. Conclusiones Principales (Análisis de Causalidad)
La clasificación de estas comunidades en **Riesgo Alto** por el modelo K-Means se debe a una clara correlación entre vulnerabilidad económica y peores indicadores nutricionales (*gradiente social de la salud*):

*   **La brecha de renta como determinante de salud:** Este clúster concentra los ingresos más bajos de España. **Extremadura** registra la menor renta per cápita del país (**13.379 €**), seguida de Canarias (14.412 €) y Murcia (14.759 €). Existe una diferencia de hasta ~9.000 € respecto a comunidades de riesgo bajo como Madrid (22.478 €).

Todas las CCAA de este grupo registran los niveles más bajos de consumo de frescos del país (siempre **<13,9**), con el mínimo en **Extremadura (13,73)** y **Castilla-La Mancha (13,81)**.

Esto contrasta con el <9,1 de Madrid o el País Vasco. Los ultraprocesados funcionan como sustitutos calóricos baratos y de larga conservación.
*   **La variable educativa:** Estas CCAA registran los menores porcentajes de estudios superiores (todos **<33%**, con Castilla-La Mancha a la cola con 29,3%). La menor formación correlaciona directamente con menor alfabetización nutricional y menor capacidad de renta.

> 📊 **Diagnóstico ejecutivo:** El riesgo nutricional en España no es una decisión individual, sino un problema de **desigualdad económica y educativa**. El sur y el centro-sur sufren un doble impacto: menores ingresos y mayor dependencia de alimentos ultraprocesados de bajo coste.
